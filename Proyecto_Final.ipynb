## Movimiento Rectilíneo Uniforme

Es aquel en donde la aceleración que tiene un cuerpo es cero, por lo que la
velocidad no varía con respecto al tiempo.

## Movimiento Rectilíneo Uniformemente Acelerado

Es aquel en el que la aceleración es constante y no nula, por lo que la velocidad cambia con respecto al tiempo.

## Movimiento Armónico Simple

Es aquel movimiento en el que la fuerza del sistema es directamente proporcional
al desplazamiento con respecto al equilibro.

## Fuerza Elástica

Fuerza que ejerce un resorte sobre un cuerpo adherido, gracias al cambio de
posición $x$ que experimenta éste. Su valor está dado por:

$$
F_e = -k \cdot x   \qquad [N] \tag{4.1}
$$ 

## Trabajo

Una fuerza $F$ realiza trabajo sobre un cuerpo cuando éste sufre un desplazamiento $d\mathbf{r}$ en un intervalo $[a, b]$ en un tiempo $dt$.

## Fuerza Conservativa

Una fuerza $F$, es conservativa cuando el trabajo de dicha fuerza se puede expresar como la diferencia del valor inicial y el valor final de una cantidad escalar $E_p(x, y, z)$, llamada energía potencial, que es función de las coordenadas de posición $x$, $y$, $z$ las cuales dependen de las dimensiones del sistema de referencia. $F$ es conservativa si cumple la siguiente condición:

$$
\begin{eqnarray}
W_{AB} &=& \int_{A}^{B}\mathbf{F}\cdot d\mathbf{r} = E_{pA}(x, y, z) - E_{pB}(x, y, z) \qquad [J] \nonumber \\
W_{AB} &=& -\Delta E_p(x, y, z) \qquad [J] \tag{4.2}
\end{eqnarray}
$$ 

## Conservación de la Energía

La energía es una cantidad física escalar definida como la capacidad para realizar un trabajo. En física clásica, la ley de conservación de la energía indica que la energía total de un sistema físico aislado permanece invariable con el tiempo.

$$
\begin{eqnarray}
&E& = E_c + E_p \qquad [J] \nonumber \\
&E& = \dfrac{1}{2}mv^2 + E_p(x, y, z) \qquad [J] \tag{4.3} \\
&E&_A = E_B \qquad [J] \nonumber
\end{eqnarray}
$$

## Resolviendo el sistema

En el punto $O$ es donde sucede el cambio de tipo de movimiento sobre la superficie horizontal. A modo de partícula, establecemos un sistema de referencia en una dimensión de tal modo que $O$ sea el origen de las coordenadas, así, el punto $d$ vendría siendo un punto negativo. Como se muestra en la siguiente figura:

Las fuerzas que actúan sobre el bloque en el trayecto son: la normal y el peso que no realizan trabajo sobre el bloque debido a que son perpendiculares al desplazamiento del bloque; y la fuerza elástica que actúa hasta el punto $O$ realiza un trabajo sobre el bloque provocándole un desplazamiento $dx$, que hallamos de la siguiente manera usando las ecuaciones (4.1) y (4.2) :

$$
\begin{eqnarray}
W_{dO} &=& \int_{-d}^{0}-k\cdot x\cdot dx = -k \int_{^-d}^{0} x \cdot dx = -\dfrac{1}{2}k \cdot (0^2-d^2 ) \qquad [J] \nonumber \\
W_{dO} &=& \dfrac{1}{2}k \cdot d^2 \qquad [J]
\end{eqnarray}
$$ 

Observemos que $W_{dO}=-Δ(\dfrac{1}{2} k \cdot x^2 )=-E_p (x)$ donde $x$ es cualquier punto en el eje de coordenadas del sistema de referencia, así, la energía potencial está dada por:

$$
E_p(x) = \dfrac{1}{2}k \cdot x^2 \qquad [J]
$$

Ahora bien, dado que la fuerza elástica es conservativa, entonces el sistema es conservativo y esto implica que la energía se conserva a medida que el bloque se mueve sobre la superficie horizontal. Aquí intervienen las ecuaciones (4.3) :

$$
\begin{eqnarray}
E_{cd} + E_p(-d) &=& E_{cO} + E_p(0) \\
\dfrac{1}{2}mv_d^2 + \dfrac{1}{2}kd^2 &=& \dfrac{1}{2}mv_0^2 + \dfrac{1}{2}kd^2 
\end{eqnarray}
$$ 

Debido a que el cuerpo se encuentra en reposo en el punto $d$, entonces su velocidad es cero, por tanto no tiene energía cinética en el punto $-d$. Así, obtenemos:

$$
\dfrac{1}{2}kd^2 = \dfrac{1}{2}mv_0^2 \qquad [J]
$$

El bloque sale entonces con una velocidad $v_0$ desde el punto $O$, y debido a que el movimiento es rectilíneo uniforme, dicha velocidad va a ser constante hasta el punto $A$, así, obtenemos una de las incógnitas del problema:

$$
V_A = \sqrt{\dfrac{k}{m}} \cdot d \qquad \left[\dfrac{m}{s}\right] \tag{5.1}
$$

Del punto $A$ al punto $B$, el bloque comienza a subir sobre el plano inclinado.	Aquí nuevamente entra la problemática de definir con respecto a qué, medir las propiedades de movimiento que describe el bloque. Nuevamente analizamos el sistema a modo de partícula, en el cual el punto $A$ toma el papel de ser el origen de coordenadas, como se muestra en la figura 5.2. Siendo así, entonces $A=0$ y $B$ vendría siendo el máximo desplazamiento del bloque sobre el plano inclinado, es decir, $B=S$. 

Aquí el sistema de referencia viene siendo de dos dimensiones debido a que el peso cuenta con dos coordenadas rectangulares.

Las coordenadas del peso en el sistema de referencia son:
+ El peso en el eje $x$, correspondiente al valor $w_x=-w \cdot sen⁡φ=-m \cdot g \cdot senφ$ $[J]$ que tiene sentido negativo debido a que el bloque se mueve en sentido positivo en el sistema de referencia.
+ El peso en el eje $y$, correspondiente al valor $w_y=-w \cdot cos⁡φ=-m \cdot g \cdot cosφ$ $[J]$ que tiene sentido negativo debido a que esta componente se encuentra apuesta al eje $y$ positivo del sistema de referencia.

Las fuerzas que actúan sobre el bloque en el trayecto sobre el plano son: la normal y el peso en el eje $y$, las cuales no realizan trabajo por ser perpendiculares al desplazamiento del bloque; y el peso en el eje $x$ que ejerce un trabajo negativo sobre el bloque cuando éste comienza a ascender desde el punto $A$ hasta el punto $B$. Su valor, aplicando las ecuaciones (4.1) y (4.2), es equivalente a:

$$
\begin{eqnarray}
W_{AB} &=& \int_{0}^{S}-mgsenφ \cdot dx = -mgsenφ \int_{^-d}^{0} dx = -mgsenφ(S-0) \qquad [J] \\
W_{AB} &=& -m \cdot g \cdot S \cdot  senφ \qquad [J]
\end{eqnarray}
$$

Observemos que $W_{AB}=-\Delta (m.g.x.senφ)=-E_p (x)$ donde $x$ es cualquier punto en el eje x de coordenadas del sistema de referencia, de modo que el peso $w_x$ es una fuerza conservativa, por tanto la energía potencial está dada por:

$$
E_p(x) = m \cdot g \cdot x \cdot  senφ \qquad [J]
$$

El sistema sigue siendo conservativo, lo que implica que la energía se sigue conservando a medida que el bloque se mueve sobre la superficie del plano inclinado.

$$
\begin{eqnarray}
E_{cA} + E_pA(0) &=& E_{cB} + E_pB(S) \\
\dfrac{1}{2}mv_A^2 + m \cdot g \cdot (0) \cdot senφ &=& \dfrac{1}{2}mv_B^2 + m \cdot g \cdot S \cdot senφ
\end{eqnarray}
$$ 

Cuando el bloque llega al punto $B$, éste se detiene tomando una velocidad igual a cero. Además, como ya tenemos el valor de la velocidad del bloque en el punto $A$ con la ecuación (5.1), reemplazando en la ecuación anterior obtenemos:

$$
\dfrac{1}{2}m \dfrac{k}{m}d^2 = m \cdot g \cdot S \cdot senφ
$$

Despejamos $S$:

$$
S = \dfrac{kd^2}{2mgsenφ} \qquad [m] \tag{5.2}
$$

En conclusión, de la ecuación (5.1) tenemos que, entre mayor sea la deformación $d$ inicial del resorte, obtendremos que el bloque tomará mayor velocidad y por ende, mayor energía cinética. De la ecuación (5.2) tenemos que para una deformación $d$ fija, a mayor ángulo $φ$ de inclinación menor es el desplazamiento del bloque sobre el plano inclinado. Sin embargo, para un ángulo $φ$ fijo, entre mayor sea la deformación $d$ del resorte, mayor va a ser el desplazamiento del bloque sobre el plano inclinado.

## Restricciones de la solución

Revisemos las condiciones que deben tener las variables de entrada $m$, $k$, $d$ y $φ$, y los valores de salida $v_A$ y $S$:
+	Todas las entradas deben ser valores positivos, la masa, la constante elástica y la deformación del resorte no pueden ser $0$.
+ El ángulo $φ$ debe estar en el intervalo de $[0, \dfrac{\pi}{2})$ en grados. Para estos valores, el $senφ$ siempre es positivo.
+	Cuando el ángulo es cero, implica que no hay un plano inclinado a considerar por lo que el máximo desplazamiento $S$ no existe.
+	Según la ecuación (5.1) y (5.2), la magnitud de la velocidad del bloque en el punto $A, v_A$, y el máximo desplazamiento $S$ sobre el plano siempre serán positivos ya que los parámetros de los que dependen son positivos.

```
def graficar2D(m, k, d, phi, s, va):
  fi = py.radians(phi)  # Convertimos el ángulo a radianes: [°] --> [Rad]
  h = s*py.sin(fi)      # Altura en [m] del plano inclinado con respecto a x
  x = s*py.cos(fi)      # Coordenada en x en [m] del plano inclinado partiendo del punto A
  enx = d+10+x          # Coordenada en x en [m] partiendo del origen

  fig = plt.figure('Movimiento del bloque', figsize=(15, 10))  # Definimos la figura que posteriormente modificaremos
  ax = plt.subplot(1, 1, 1)   # Establecemos un eje de coordenadas

  plt.xlabel('x[m]', fontsize=15, color='white')  # Labels de las coordenadas
  plt.ylabel('y[m]', fontsize=15, color='white')
  plt.xlim(-d-20, enx+5)  # Límites de los ejes en [m] en el eje x
  plt.ylim(-15, h+20)     # Límites de los ejes en [m] en el eje y

  # Agregamos algunas líneas negras
  plt.axhline(y=0, color='black', linewidth=2)
  # Las siguientes instrucciones unen dos puntos con una línea
  plt.plot([-d-19, -d-19], [20, 0], color='black', linewidth=2)   # En el eje y: plot([x2, x1], [y2, y1]
  plt.plot([enx, enx], [h, 0], linestyle='--', color='b', linewidth=2)  # Agregamos la altura del plano inclinado
  plt.plot([enx, d+10], [h, 0], 'b')  # Agregamos la hipotenusa del plano inclinado

  # Dibujamos el resorte
  xf = py.linspace(-d-19, -d-5, 360)
  fx = 1.5*py.sin(6*xf)+2.5
  plt.plot(xf, fx)

  # Agregamos los bloques con la librería patches
  bloque = pat.Rectangle((-d-5, 0), 5, 5, linestyle='--', edgecolor='r', facecolor='none', linewidth=2.2)
  bloque2 = pat.Rectangle((enx-5*py.cos(fi), h-5*py.sin(fi)), 5, 5, angle=phi, facecolor='red', linewidth=3)
  bloque3 = pat.Rectangle((d+5, 0), 5, 5, linestyle='--', edgecolor='r', facecolor='none', linewidth=2.2)
  # Mostramos los bloques en el plano de coordenadas
  ax.add_patch(bloque)
  ax.add_patch(bloque2)
  ax.add_patch(bloque3)

  txt = str(m)+' kg'
  plt.annotate(txt,    # Ésta anotación agrega una flecha con la masa apuntando al bloque
              xy=(-d-3,5), xycoords='data',   # Las coordenadas xy están dadas en cuanto a los datos
              xytext=(+0, +30), textcoords='offset points', fontsize=15, 
              arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=.2", linewidth=1.2))
  
  # Agregamos una flecha estilo Fancy para el ángulo de inclinación del plano inclinado
  if x<=20:   # Este condicional es para establecer las coordenadas del texto
    a = x/3 + 2
  else: a = x/4 + x/6
  xA = pat.FancyArrowPatch((d+10+x/3, 0), (d+10+x/3, x/3*py.tan(fi)), color='b', 
                          arrowstyle="fancy", connectionstyle="arc3,rad=.3", mutation_scale=20)
  txt = str(phi)+'°'
  plt.text(d+10+a, x/6*py.tan(fi), txt, fontsize=15)
  ax.add_artist(xA) # Agregamos al plano de coordenadas la flecha Fancy

  # Agregamos una anotación para la deformación del resorte
  txt = str(d)+' m'
  if h<=20:  # Condicional para ubicar el texto de la deformación del resorte
    yd=15
    yt=17
  else:
    yd = h/2.5
    yt = h/2.2
  plt.annotate('', xy=(-d, yd), xycoords='data', xytext=(0, yd), 
              arrowprops=dict(arrowstyle='|-|', linewidth=1.2, linestyle='--'), va='center' )
  plt.text(-d, yt, txt, fontsize=15)  # Mostramos el texto

  # Ecuación de la recta del plano inclinado para ubicar el texto del desplazamiento (inversa)
  x1 = (enx-d-10)/2  +d+10 -7*py.sin(fi) -2   # Ecuación: y(enx-d-10)/2 +d+10 = x
  y1 = h/2 + 7*py.cos(fi)
  txt = str(s)+' m'
  plt.annotate(txt, xy=(d+10-7*py.sin(fi), 7*py.cos(fi)), xycoords='data', xytext=(x1, y1), fontsize=15,
              arrowprops=dict(arrowstyle='-|>', linestyle='dashdot', linewidth=2))
  plt.annotate(txt, xy=(enx-7*py.sin(fi), h+7*py.cos(fi)+1), xycoords='data', xytext=(x1, y1), fontsize=15,
              arrowprops=dict(arrowstyle='-|>', linestyle='dashdot', linewidth=2))
  
  # Texto para mostrar la constante elástica del resorte
  txt = str(k)+' N/m'
  plt.text(-d-18, -5, txt, fontsize=15)

  # Anotación que muestra la velocidad del bloque en el punto A del plano horizontal
  txt = str(va)+' m/s'
  plt.annotate(txt, xy=(d+10, 0), xycoords='data', xytext=(-20, -50), textcoords='offset points', fontsize=15,
              arrowprops=dict(arrowstyle="-|>", linewidth=1.5))

  # Agregamos la referencia de la imagen
  msg = 'figura %d: Resultados de la prueba' %(i+1)
  plt.figtext(0.2, 0.02, msg + ' #%d \n' %(i+1),transform=plt.gcf().transFigure,fontsize=20)

  plt.show()  # Por último mostramos la gráfica resultante
```
